### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda2.head()
```

#1. Análisis de facturación



```python
#Ingreso total por cada tienda
ingreso_tienda1 =sum(tienda['Precio'])
ingreso_tienda2 =tienda2['Precio'].sum()
ingreso_tienda3 =sum(tienda3['Precio'])
ingreso_tienda4 =sum(tienda4['Precio'])
print(f"Ingreso total tienda 1 es : {ingreso_tienda1}")
print(f"Ingreso total tienda 2 es : {ingreso_tienda2}")
print(f"Ingreso total tienda 3 es : {ingreso_tienda3}")
print(f"Ingreso total tienda 4 es : {ingreso_tienda4}")
```

```python
import matplotlib.pyplot as plt

# Datos de ejemplo
tienda_ingreso = ["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"]
total_tiendas = [ingreso_tienda1, ingreso_tienda2, ingreso_tienda3, ingreso_tienda4]

# Calcular el total
total_ingresos = sum(total_tiendas)

# Función para mostrar porcentaje y subtotal
def func_autopct(pct, valores):
    # Subtotal de cada porción
    valor = int(round(pct/100.*sum(valores)))
    # Mostrar porcentaje y subtotal debajo
    return f"{pct:.1f}%\n{valor:,}".replace(".", ",")

# Crear gráfico circular
fig, ax = plt.subplots()
ax.pie(
    total_tiendas,
    labels=tienda_ingreso,
    autopct=lambda pct: func_autopct(pct, total_tiendas),
    startangle=90,
    colors=["skyblue","lightgreen","salmon","gold"]
)

# Mostrar el total en el centro
plt.text(0, 0, f"Total:\n{total_ingresos:,}".replace(".", ","),
         ha="center", va="center", fontsize=12, fontweight="bold")

# Título
ax.set_title("Ingresos por Tienda", fontsize=14)

plt.tight_layout()
plt.show()
```

##La distribución de categorías de productos

```python
import pandas as pd
import matplotlib.pyplot as plt
#  Convertir cada conteo en DataFrame y renombrar columnas
df1 = vc_tienda1.rename("Tienda 1")
df2 = vc_tienda2.rename("Tienda 2")
df3 = vc_tienda3.rename("Tienda 3")
df4 = vc_tienda4.rename("Tienda 4")
#  Unir todos en una sola tabla comparativa
df_categorias = pd.concat([df1, df2, df3, df4], axis=1).fillna(0).astype(int)
#  Graficar barras múltiples
ax = df_categorias.plot(kind="bar", figsize=(10,6), colormap="Set2")

# Personalización ejecutiva
ax.set_title("Cantidad de Productos Vendidos por Categoría", fontsize=14)
ax.set_xlabel("Categoría de Producto", fontsize=12)
ax.set_ylabel("Cantidad de Productos", fontsize=12)
ax.legend(title="Tiendas", bbox_to_anchor=(1.05, 1), loc="upper left")

# Opcional: ajustar rango del eje Y para que se noten mejor las diferencias
ax.set_ylim(150, 500)

plt.tight_layout()
plt.show()

# 6. Mostrar la tabla comparativa en consola
print("Tabla comparativa de cantidad de productos por categoría y tienda:")
print(df_categorias)
```
