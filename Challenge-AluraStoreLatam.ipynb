### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
#Ingreso total por cada tienda
ingreso_tienda1 =sum(tienda['Precio'])
ingreso_tienda2 =tienda2['Precio'].sum()
ingreso_tienda3 =sum(tienda3['Precio'])
ingreso_tienda4 =sum(tienda4['Precio'])
print(f"Ingreso total tienda 1 es : {ingreso_tienda1}")
print(f"Ingreso total tienda 2 es : {ingreso_tienda2}")
print(f"Ingreso total tienda 3 es : {ingreso_tienda3}")
print(f"Ingreso total tienda 4 es : {ingreso_tienda4}")
```

# 2. Ventas por categoría

```python
vc_tienda1 = tienda['Categoría del Producto'].value_counts()
vc_tienda2 = tienda2['Categoría del Producto'].value_counts()
vc_tienda3 = tienda3['Categoría del Producto'].value_counts()
vc_tienda4 = tienda4['Categoría del Producto'].value_counts()
```

```python
#Cantidad de producto Tienda 1
print(vc_tienda1)
```

```python
#Cantidad de producto Tienda 4
print(vc_tienda4)
```

```python
#Ventas por categoría Tienda
venta_t1=tienda.groupby('Categoría del Producto')[['Precio']].sum().sort_values("Precio",ascending=False) # por usar doble corchete es una DataFrame
venta_t2=tienda2.groupby('Categoría del Producto')['Precio'].sum().sort_values(ascending=False)
venta_t3=tienda3.groupby('Categoría del Producto')['Precio'].sum().sort_values(ascending=False)
venta_t4=tienda4.groupby('Categoría del Producto')['Precio'].sum().sort_values(ascending=False)
```

```python
#Ventas por categoría Tienda 1
print(venta_t1)
```

```python
#Ventas por categoría TIENDA 2
print(venta_t2)
```

```python
#Ventas por categoría TIENDA 3
print(venta_t3)
```

```python
#Ventas por categoría TIENDA 4
print(venta_t4)
```

# 3. Calificación promedio de la tienda


```python
promt1 = tienda['Calificación'].mean().round(2)
promt2 = tienda2['Calificación'].mean().round(2)
promt3 = tienda3['Calificación'].mean().round(2)
promt4 = tienda4['Calificación'].mean().round(2)
promedios = {
    "Tienda 1": promt1,
    "Tienda 2": promt2,
    "Tienda 3": promt3,
    "Tienda 4": promt4
}
df_promedios = pd.DataFrame(list(promedios.items()),columns=["Tienda","Calificación Prom de clientes"])
print(df_promedios)
```

# 5. Envío promedio por tienda

```python
prom_envio_t1=round(tienda['Costo de envío'].mean(), 2)
prom_envio_t2=round(tienda2['Costo de envío'].mean(), 2)
prom_envio_t3=round(tienda3['Costo de envío'].mean(), 2)
prom_envio_t4=round(tienda4['Costo de envío'].mean(), 2)
promediosEnvio = {
    "Tienda 1": prom_envio_t1,
    "Tienda 2": prom_envio_t2,
    "Tienda 3": prom_envio_t3,
    "Tienda 4": prom_envio_t4

}
df_envio = pd.DataFrame(list(promediosEnvio.items()),columns=["Tienda","Promedio de Envío"])
print(df_envio)
```
