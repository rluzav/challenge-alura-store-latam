### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda2.head()
```

```python
import matplotlib.pyplot as plt

# Datos de ejemplo
tienda_ingreso = ["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"]
total_tiendas = [ingreso_tienda1, ingreso_tienda2, ingreso_tienda3, ingreso_tienda4]

# Calcular el total
total_ingresos = sum(total_tiendas)

# Función para mostrar porcentaje y subtotal
def func_autopct(pct, valores):
    # Subtotal de cada porción
    valor = int(round(pct/100.*sum(valores)))
    # Mostrar porcentaje y subtotal debajo
    return f"{pct:.1f}%\n{valor:,}".replace(".", ",")

# Crear gráfico circular
fig, ax = plt.subplots()
ax.pie(
    total_tiendas,
    labels=tienda_ingreso,
    autopct=lambda pct: func_autopct(pct, total_tiendas),
    startangle=90,
    colors=["skyblue","lightgreen","salmon","gold"]
)

# Mostrar el total en el centro
plt.text(0, 0, f"Total:\n{total_ingresos:,}".replace(".", ","),
         ha="center", va="center", fontsize=12, fontweight="bold")

# Título
ax.set_title("Ingresos por Tienda", fontsize=14)

plt.tight_layout()
plt.show()
```

##Producto más vendido y menos vendido

```python
import matplotlib.pyplot as plt

# Extraer los valores (cantidades) de productos más y menos vendidos
max_t1 = prod_mas_t1.values[0]
max_t2 = prod_mas_t2.values[0]
max_t3 = prod_mas_t3.values[0]
max_t4 = prod_mas_t4.values[0]

min_t1 = prod_menos_t1.values[-1]  # El último (menos vendido)
min_t2 = prod_menos_t2.values[-1]
min_t3 = prod_menos_t3.values[-1]
min_t4 = prod_menos_t4.values[-1]

# Extraer nombres de productos
nombre_mas_t1 = prod_mas_t1.index[0]
nombre_mas_t2 = prod_mas_t2.index[0]
nombre_mas_t3 = prod_mas_t3.index[0]
nombre_mas_t4 = prod_mas_t4.index[0]

nombre_min_t1 = prod_menos_t1.index[-1]
nombre_min_t2 = prod_menos_t2.index[-1]
nombre_min_t3 = prod_menos_t3.index[-1]
nombre_min_t4 = prod_menos_t4.index[-1]

# Preparar datos con nombres
datos = {
    'Tienda 1': [max_t1, min_t1],
    'Tienda 2': [max_t2, min_t2],
    'Tienda 3': [max_t3, min_t3],
    'Tienda 4': [max_t4, min_t4]
}

df_grafico = pd.DataFrame(datos, index=['Más vendido', 'Menos vendido'])

# Crear gráfico
ax = df_grafico.T.plot(kind='bar', figsize=(12, 7), color=['green', 'red'], alpha=0.7)
plt.title('Comparación de Productos Más y Menos Vendidos por Tienda', fontsize=14, fontweight='bold')
plt.xlabel('Tiendas', fontsize=12)
plt.ylabel('Cantidad de ventas', fontsize=12)
plt.xticks(rotation=45)
plt.legend(title='Producto')
plt.grid(axis='y', alpha=0.3)

# Lista de nombres por tienda
nombres_productos = [
    [nombre_mas_t1, nombre_min_t1],
    [nombre_mas_t2, nombre_min_t2],
    [nombre_mas_t3, nombre_min_t3],
    [nombre_mas_t4, nombre_min_t4]
]

# Agregar etiquetas con nombres de productos
for i, container in enumerate(ax.containers):
    labels = [nombres_productos[j][i] for j in range(len(nombres_productos))]
    ax.bar_label(container, labels=labels, fontsize=8, fontweight='bold', padding=3)

plt.tight_layout()
plt.show()
```
